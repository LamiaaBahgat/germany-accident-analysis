```python
# Section 2 — Accident Distribution by Year / State / Category
# Interactive Icicle chart — click to drill down into states and categories
# Abschnitt 2 — Unfallverteilung nach Jahr / Bundesland / Kategorie
# Interaktives Icicle-Diagramm — zum Aufklappen klicken

import plotly.graph_objects as go
import pandas as pd

# Group data by year, state and category
# Daten nach Jahr, Bundesland und Kategorie gruppieren
df_counts = df.groupby(['year', 'state', 'category']).size().reset_index(name='accident_count')

levels = ['category', 'state', 'year']
value_column = 'accident_count'

def build_hierarchical_dataframe(df, levels, value_column):
    df_list = []
    for i, level in enumerate(levels):
        df_tree = pd.DataFrame(columns=['id', 'parent', 'value', 'color'])
        dfg = df.groupby(levels[i:])[value_column].sum().reset_index()

        # Make unique IDs by joining all levels from current to top
        dfg['unique_id'] = dfg[levels[i:]].astype(str).agg(' | '.join, axis=1)
        df_tree['id'] = dfg['unique_id']
        df_tree['value'] = dfg[value_column]
        df_tree['color'] = dfg[value_column]

        if i < len(levels) - 1:
            dfg['parent_id'] = dfg[levels[i+1:]].astype(str).agg(' | '.join, axis=1)
            df_tree['parent'] = dfg['parent_id']
        else:
            df_tree['parent'] = 'Germany'

        df_list.append(df_tree)

    # Root node — Germany total
    # Wurzelknoten — Deutschland gesamt
    total = pd.Series(dict(
        id='Germany',
        parent='',
        value=df[value_column].sum(),
        color=df[value_column].sum()
    ), name=0)
    df_list.append(total)

    df_all_trees = pd.concat(df_list, ignore_index=True)
    return df_all_trees

df_all_trees = build_hierarchical_dataframe(df_counts, levels, value_column)

# Color scale: yellow → orange → red (low → high accidents)
# Farbskala: Gelb → Orange → Rot (wenig → viele Unfälle)
custom_colorscale = [
    [0.0, '#f1c21b'],
    [0.5, '#ff832b'],
    [1.0, '#da1e28'],
]

fig = go.Figure()

fig.add_trace(go.Icicle(
    labels=df_all_trees['id'],
    parents=df_all_trees['parent'],
    values=df_all_trees['value'],
    branchvalues='total',
    marker=dict(
        colors=df_all_trees['color'],
        colorscale=custom_colorscale,
        showscale=True,
        colorbar=dict(title='Accidents')
    ),
    hovertemplate='<b>%{label}</b><br>Accidents: %{value:,}<extra></extra>',
    maxdepth=2,
    tiling=dict(
        orientation='v',
        pad=2
    ),
    textfont=dict(size=10),
    pathbar=dict(visible=True)
))

fig.update_layout(
    title=dict(
        text='🇩🇪 Accidents by Year / State / Category<br><sup>Click to drill down | Zum Aufklappen klicken</sup>',
        x=0.5,
        xanchor='center'
    ),
    margin=dict(t=80, l=25, r=25, b=25),
    width=1400,
    height=700,
    paper_bgcolor='white',
)

fig.show()
```

```python
#Section 3 —  Weather analysis
#Schritt 3 — Analyse der Wetterbedingungen

# Step 1 — Load data - 2021
# Schritt 1 — Daten laden-2021

weather_2021 = pd.read_csv(os.path.join(WEATHER_DIR, 'combined_2021_hourly_weather_data.csv'), low_memory=False)
```

```python
weather_2022 = pd.read_csv(os.path.join(WEATHER_DIR, 'combined_2022_hourly_weather_data.csv'), low_memory=False)
```

```python
weather_2023 = pd.read_csv(os.path.join(WEATHER_DIR, 'combined_2023_hourly_weather_data.csv'), low_memory=False)
```

```python
# Combine all weather years
# Zusammenführung aller Wetterjahre

weather = pd.concat([weather_2021, weather_2022, weather_2023], ignore_index=True)
print(f"✅ Total weather rows: {len(weather)}")
```

```python
# Step 2 — Fix timezone
# Schritt 2 — Zeitzone anpassen
# Convert timezone UTC → Berlin
# Zeitzone von UTC in Berliner Zeit umwandeln

weather['date'] = pd.to_datetime(weather['date'], utc=True)
weather['date'] = weather['date'].dt.tz_convert('Europe/Berlin')
```

```python
# Extract date and hour for joining
weather['join_date'] = weather['date'].dt.strftime('%Y-%m-%d')
weather['join_hour'] = weather['date'].dt.hour

# Round coordinates
weather['lat_round'] = weather['latitude'].round(2)
weather['lon_round'] = weather['longitude'].round(2)

print(f"✅ Timezone converted!")
print(weather[['date', 'join_date', 'join_hour', 'lat_round', 'lon_round']].head())
```

```python
# Step 3 - Prepare accidents for joining
# Schritt 3 - Unfalldaten für den Join vorbereiten

df['join_date'] = pd.to_datetime(df['estimated_date']).dt.strftime('%Y-%m-%d')
df['join_hour'] = df['hour'].astype(int)
df['lat_round'] = df['latitude'].round(2)
df['lon_round'] = df['longitude'].round(2)

print(df[['estimated_date', 'join_date',
                  'join_hour', 'lat_round',
                  'lon_round']].head())
```

```python
# Step 4 — Merge using DuckDB SQL
# Schritt 4 — Daten mit DuckDB SQL zusammenführen

merged = duckdb.query("""
    SELECT
        d.*,
        w.temperature_2m,
        w.rain,
        w.snowfall,
        w.wind_speed_10m,
        w.wind_direction_10m,
        w.weather_code AS weather_code_hourly,
        w.cloud_cover,
        w.shortwave_radiation
    FROM df d
    LEFT JOIN weather w
        ON d.join_date = w.join_date
        AND d.join_hour = w.join_hour
        AND d.lat_round = w.lat_round
        AND d.lon_round = w.lon_round
""").df()

print(f"\n✅ Merged rows: {len(merged)}")
print(f"✅ Columns: {merged.columns.tolist()}")

# Check how many accidents got weather data
matched = merged['rain'].notna().sum()
print(f"✅ Accidents with weather data: {matched} ({matched/len(merged)*100:.1f}%)")
```

```python
# 8.2 — Helper Functions
# Hilfsfunktionen


def ask_gemma(prompt):
    """Send question to Gemma 4 via Ollama"""
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "gemma4:e4b",
            "prompt": prompt,
            "stream": False
        }
    )
    return response.json()["response"]


def ask_and_save(prompt, filename):
    """Ask Gemma and save response — loads from file if already asked"""
    filepath = os.path.join(BASE_DIR, 'gemma_responses', filename)

    # Create folder if not exists
    os.makedirs(os.path.join(BASE_DIR, 'gemma_responses'), exist_ok=True)

    # Load if already saved
    if os.path.exists(filepath):
        print(f"✅ Loading saved: {filename}")
        with open(filepath, 'r') as f:
            return f.read()

    # Ask Gemma
    print(f"🤖 Asking Gemma 4... (may take 5 minutes)")
    print(f"💾 Will save to: {filename}")
    response = ask_gemma(prompt)

    # Save
    with open(filepath, 'w') as f:
        f.write(response)
    print(f"✅ Saved!")

    return response
```

```python
# ---- 8.3 Weather Impact Analysis ----
# Wettereinfluss Analyse

weather_stats = merged.groupby('category').agg(
    total=('category', 'count'),
    avg_rain=('rain', 'mean'),
    avg_temp=('temperature_2m', 'mean'),
    avg_wind=('wind_speed_10m', 'mean'),
    avg_cloud=('cloud_cover', 'mean'),
    rainy_pct=('rain', lambda x: (x > 0).mean() * 100)
).round(2)

response_weather = ask_and_save(f"""
You are a road safety expert analyzing German accident data 2021-2023.

Weather conditions per accident severity:
{weather_stats.to_string()}

Total accidents: {len(merged):,}
Fatal accidents: {len(merged[merged['category']=='killed']):,}

Analyze:
1. What weather conditions are most associated with fatal accidents?
2. Why do most accidents happen in mild/dry conditions?
3. Role of wind speed in accident severity?
4. Recommendations to reduce fatal accidents?

Be concise. Answer in English then German.
""", "gemma_weather.txt")

print("🌤️ Weather Analysis / Wetteranalyse")
print("=" * 50)
print(response_weather)
```

```python
# ---- 8.4 Location Analysis ----
location_stats = merged.groupby('state').agg(
    total=('category', 'count'),
    fatal=('category', lambda x: (x == 'killed').sum()),
    fatal_pct=('category', lambda x: (x == 'killed').mean() * 100)
).round(2).sort_values('total', ascending=False)

response_location = ask_and_save(f"""
You are a road safety expert analyzing German accident data 2021-2023.

Accidents by state (sorted by total):
{location_stats.to_string()}

Analyze:
1. Why does Nordrhein-Westfalen have the most accidents?
2. Which states have highest fatal accident percentage?
3. What regional factors explain these differences?

Be concise. Answer in English then German.
""", "gemma_location.txt")

print("📍 Location Analysis / Standortanalyse")
print("=" * 50)
print(response_location)
```

```python
# ---- 8.5 Time Pattern Analysis ----
time_stats = merged.groupby(['day', 'hour'])['category'].apply(
    lambda x: (x == 'killed').sum()
).reset_index(name='fatal_count')

most_dangerous = time_stats.nlargest(5, 'fatal_count')

# All accidents by day
all_by_day = merged.groupby('day').size().reset_index(name='total')

# Fatal by day
fatal_by_day = merged[merged['category']=='killed'].groupby('day').size().reset_index(name='fatal')

# Combine
day_analysis = all_by_day.merge(fatal_by_day, on='day')
day_analysis['fatal_pct'] = (day_analysis['fatal'] / day_analysis['total'] * 100).round(2)

response_time = ask_and_save(f"""
You are a road safety expert analyzing German accident data 2021-2023.

Accidents by day of week:
{day_analysis.sort_values('total', ascending=False).to_string()}

Top 5 most deadly day/hour combinations:
{most_dangerous.to_string()}

Key observations:
- Friday has MOST accidents (146,410) but lower fatal rate (0.82%)
- Sunday has FEWEST accidents (81,594) but HIGHEST fatal rate (1.27%)
- Monday 15:00 is single most deadly hour (112 deaths)

Analyze:
1. Why does Sunday have highest fatal rate despite fewer accidents?
2. Role of drunk/impaired driving on weekends?
3. Why is Monday 15:00 most deadly single hour?
4. How can authorities target Sunday and Monday dangers differently?

Be concise. Answer in English then German.
""", "gemma_time.txt")

print("⏰ Time Analysis / Zeitanalyse")
print("=" * 50)
print(response_time)
```

```python
# ---- 8.6 Final Recommendations ----
response_recommendations = ask_and_save(f"""
You are a road safety expert for Germany's Vision Zero initiative.

Summary of our analysis of {len(merged):,} accidents 2021-2023:

LOCATION:
- Most accidents: Nordrhein-Westfalen (density/urban problem)
- Highest fatal rate: Sachsen-Anhalt 1.61%, Brandenburg 1.24% (rural problem)

TIME PATTERNS:
- Most accidents: Friday (146,410 total)
- Most DEADLY day: Sunday (1.27% fatal rate) — suspected drunk driving
- Most deadly single hour: Monday 15:00 (112 deaths)
- Weekend fatal rate much higher than weekdays

WEATHER:
- Cold temperatures increase fatal risk
- Wind speed not a significant factor
- Most accidents happen in mild/dry conditions (high traffic volume)

FATAL ACCIDENTS:
- Total: {len(merged[merged['category']=='killed']):,}
- Rate: {len(merged[merged['category']=='killed'])/len(merged)*100:.1f}%

Based on this complete analysis, provide:
1. Top 5 actionable recommendations for German road safety
2. Different strategies for West vs East Germany
3. How to target Sunday drunk driving specifically
4. How to reduce Monday afternoon accidents
5. How does this align with Vision Zero goals?

Be specific and data-driven. Answer in English then German.
""", "gemma_recommendations.txt")

print("💡 Recommendations / Empfehlungen")
print("=" * 50)
print(response_recommendations)
```

# 🔍 Key Insights / Wichtigste Erkenntnisse

## 🌤️ Weather Impact
- Most accidents happen in **mild/dry conditions** — not bad weather
- **Cold temperature** (avg 12.4°C) correlates with fatal accidents
- **Wind speed** has minimal effect on accident severity
- Rain percentage only marginally higher for fatal accidents

## 📍 Location — Two Different Problems
- **West Germany** (NRW, Bayern): Highest accident COUNT
  → Population density + urban complexity
- **East Germany** (Sachsen-Anhalt 1.61%, Brandenburg 1.24%): Highest FATAL rate
  → Rural infrastructure + higher speed roads

## ⏰ Time Patterns — The Surprising Truth
- ❌ Everyone thinks Friday is most dangerous
- ✅ **Sunday is actually most deadly** (1.27% fatal rate)
  → Suspected drunk/impaired driving
  → Leisure drivers less focused than commuters
- **Monday 15:00** is single most deadly hour (112 deaths)
  → "Sunday Scaries" stress + fatigue effect

## 🚗 Scale of Analysis
- **782,642** accidents analyzed
- **3 years** of data (2021-2023)
- **99.7%** weather data matched
- **7,952** fatal accidents

- Different strategies needed for West vs East Germany
- Sunday: Alcohol intervention + saturation policing
- Monday: Fatigue awareness campaigns 14:00-17:00
- NRW: Engineering solutions (infrastructure)
